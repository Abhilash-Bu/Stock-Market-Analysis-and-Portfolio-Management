```python
s=df['Close Price']
s=s.pct_change()  #calculating the percentage of close prices 
#print(s)
s=s.fillna(0)#filling the NaN values with zeros
s=s*100
#adding a new column to existing dataframe is done by insert() or assign() function
df.insert(6,'Day_Perc_Change',s)
#df.assign('Day_Perc_Change':[s])
```

```python
df.insert(7,'Trend','NaN') #creating a blank column with NaN values
```

```python
del df['Trend'] #deleting an exsisting column
```

```python
#df[((df['Day_Perc_Change']<0.5) & (df['Day_Perc_Change']>-0.5))]
df.loc[(df['Day_Perc_Change'] > -0.5) & (df['Day_Perc_Change'] < -0.5), 'Trend'] = 'No change'
df.loc[(df['Day_Perc_Change'] > 0.5) & (df['Day_Perc_Change'] < 1), 'Trend'] = 'Slight positive'
df.loc[(df['Day_Perc_Change'] > -1) & (df['Day_Perc_Change'] < -0.5), 'Trend'] = 'Slight negative'
df.loc[(df['Day_Perc_Change'] > 1) & (df['Day_Perc_Change'] < 3), 'Trend'] = 'Positive'
df.loc[(df['Day_Perc_Change'] > -3) & (df['Day_Perc_Change'] < -1), 'Trend'] = 'Negative'
df.loc[(df['Day_Perc_Change'] > 3) & (df['Day_Perc_Change'] < 7), 'Trend'] = 'Among top gainers'
df.loc[(df['Day_Perc_Change'] > -7) & (df['Day_Perc_Change'] < -3), 'Trend'] = 'Among top losers'
df.loc[(df['Day_Perc_Change'] > 7), 'Trend'] = 'Bull run'
df.loc[(df['Day_Perc_Change'] < -7), 'Trend'] = 'Bear drop'
```

```python
grouped_year=df.groupby('Year') #grouping based on years
#grouped_year.first() #this command is used to get the first values of each group
grouped_2017=grouped_year.get_group(2017) #obtaining 2017 year data
grouped_2018=grouped_year.get_group(2018) #obtaining 2018 year data
grouped_2019=grouped_year.get_group(2019) #obtaining 2019 year data
#df.groupby
```

```python
#Finding VWAP for each month of year 2017
grouped_2017_mean=grouped_2017.groupby('Month').mean() #finding the mean of all the values to find VWAP
grouped_2017_mean.insert(9,'VWAP','NaN')
for i in range(grouped_2017_mean.shape[0]):
    grouped_2017_mean.iat[i,9]=grouped_2017_mean.iat[i,6]/grouped_2017_mean.iat[i,5]
grouped_2017_mean
```
